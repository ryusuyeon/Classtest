```python
import pandas as pd

num_dataframes = 100 # 관직 개수을 넣으면 됨.

base_cols = [0, 1, 2, 3, 4]
use_cols = []

start = 5
end = 324 # 관직에 개수 따라 정해지는 수. 종친부는 18개의 관직이 존재함. 기본 식은 ' end = 4n+4'. 이며 n에 관직에 개수를 넣으면 됨. 
         # 종친부는 18개 관직이므로 end = 4 * 18 + 4 = 76 (기본 엑셀 파일에서 76번째 열까지 사용한다는 의미임)
step = 4

while start <= end:
    use_cols.append(base_cols + list(range(start, start + step)))
    start += step

new_columns = ['왕', '재위', '연도', '월', '일', '이름', '유형', '출처', '메모']
insertions = [(0, 'AGE_CODE', ' '), (1, 'KING_CODE', ' '), (5, '윤', ' '), (9, '이름(한자)', ' ')]
df_names = ['경기관찰사', '경기도도관찰사', '경기좌도관찰출척사', '경기좌도도관찰사', '경기좌도관찰사', '경기좌도안렴사'
, '경기우도관찰출척사', '경기우도도관찰사', '경기우도관찰사', '경기우도안렴사', '경기좌우도관찰출척사', '경기좌우도도관찰사'
, '경기도관찰출척사병마도절제사', '여주목사', '여주군수', '여주현감', '파주목사', '양주목사', '양주도호부사', '양주부사'
, '지양주사', '양주군수', '부평도호부사', '부평부사', '장단도호부사', '장단부사', '장단부사겸방어사', '장단군수', '장단현령'
, '장단현감', '남양도호부사', '남양부사', '판남양도호부사', '판남양도회부사', '남양군수', '남양현감', '통진부사'
, '통진군수', '통진현감', "이천도호부사", '이천부사', '이천현감', '교동부사', '판교동현사', '교동군수'
, '교동현감', '인천도호부사', '인천부사', '인천부윤', '인천현감', '죽산도호부사', '죽산부사', '죽산현감'
, '양근군수', '양근현감', '고양군수', '고양현령', '고양현감', '안산군수', '안산현감', '김포군수', '김포현령', '김포현감'
, '삭녕군수', '삭녕현감', '교하군수', '교하현감', '안성군수', '가평군수', '가평현감', '마전군수', '마전현감'
, '마전감무', '영평군수', '영평현령', '영평현감', '경기도도사', '경기도사', '경기우도도사', '겸경기도사', '여주판관', '영평판관'
, '화성판관', '용인현령', '용인현감', '진위현령', '진위현감', '양천현령', '양천현감', '시흥현령', '시흥군수', '시흥현감'
, '경기도찰방', '경기도찰방2', '경기도해도찰방', '경기우도찰방', '경기정역찰방', '경기좌도찰방', '경기찰방', '경안도찰방'] # 파일에 있는 관직 이름 넣기

combined_df = pd.DataFrame()

for i in range(num_dataframes):
    df = pd.read_excel('경기도-1.xlsx', usecols=use_cols[i]) # 기본 파일 이름으로 변경
    df.columns = new_columns
    df = df.dropna(subset=['이름'])
    for insertion in insertions:
        df.insert(insertion[0], insertion[1], insertion[2], insertion[2])

    df['이름(한자)'] = df['이름'].apply(lambda x: x.split('(')[1] if '(' in x else '')

    columns_to_clean = ['왕', '이름', '이름(한자)']

    for column in columns_to_clean:
        if column != '이름(한자)':
            df[column] = df[column].str.replace(pat=r'[^\w]', repl=r'', regex=True)
            df[column] = df[column].str.replace(pat=r'[一-龥]', repl=r'', regex=True)
        else:
            df[column] = df[column].str.replace(pat=r'[^\w]', repl=r'', regex=True)

    df.insert(10, '관직명', df_names[i])

    combined_df = combined_df.append(df) 

output_filename = '경기도1.xlsx' # 최종 저장 파일 이름 지정

with pd.ExcelWriter(output_filename) as writer:
    combined_df.to_excel(writer, sheet_name='Combined', index=False)
```
